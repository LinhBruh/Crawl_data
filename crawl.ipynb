```python
import requests
import json
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import os
```

```python
base_url = "https://cellphones.com.vn/"
headers = {"User-Agent" : "Mozilla/5.0"}

response = requests.get(base_url, headers=headers)
html = response.text
```

```python
soup = BeautifulSoup(html,"html.parser")
soup
```

```python
soup = BeautifulSoup(html, "html.parser")

assets = {
    "images": [urljoin(base_url, img["src"]) for img in soup.find_all("img") if img.get("src")],
    "scripts": [urljoin(base_url, s["src"]) for s in soup.find_all("script") if s.get("src")],
    "css": [urljoin(base_url, l["href"]) for l in soup.find_all("link", rel="stylesheet") if l.get("href")],
    "videos": [urljoin(base_url, v["src"]) for v in soup.find_all("video") if v.get("src")],
    "iframes": [urljoin(base_url, f["src"]) for f in soup.find_all("iframe") if f.get("src")],
    "links": [urljoin(base_url, a["href"]) for a in soup.find_all("a") if a.get("href")],
    "audios" : [urljoin(base_url,au["src"]) for au in soup.find_all("media") if au.get("src")]
}
```

```python
assets["links"]
categories_links= [link for link in assets["links"] if link.endswith('.html') ]
```
